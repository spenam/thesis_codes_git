```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numeros=[6,8,10,16,20,26,30,36,40]
x=[]
p=[]
t=[]
for i in numeros:
    x.append(np.load("x{}.npy".format(i)))
    p.append(np.load("p{}.npy".format(i)))
    t.append(np.load("t{}.npy".format(i)))
    
def Hsys(x,p):
    return p*p/2. -0.1*x*x + 0.0025*x*x*x*x
```

```python
def Hsys(x,p):
    return p*p/2. -0.1*x*x + 0.0025*x*x*x*x
vfunc = np.vectorize(Hsys)
```
